# #13: 3D Plotting

- Membuat 3D plotting dengan Matplotlib

## Import modules

- `%matplotlib inline` untuk memanggil magic comment. Fungsinya untuk meredirect hasil output dari matplotlib ke tampilan jupyer   notebook di halaman notebook yang sama
- Fungsi `%matplotlib notebook` sama seperti `%matplotlib inline`, tetapi sewaktu mengaktifkan `%matplotlib notebook` ini kita bisa berinteraksi dengan 3D plotting yang kita hasilkan
- Kode di bawah ini untuk memberi tahu berapa versi matplotlib dan numpy kalian

```python
# %matplotlib inline
%matplotlib notebook
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

## 3D Axes

- Memanggil object 3D harus membangun sendiri object figure dan object axes 

```python
fig = plt.figure()             # membentuk object figure
ax = plt.axes(projection='3d') # membentuk object axes 3D

plt.show()
```

## 3D Line Plot

- Tiga variabel di bawah akan di plotting pada area 3D di atas

```python
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
```

```python
fig = plt.figure()             
ax = plt.axes(projection='3d')

ax.plot(x, y, z, color='red') # optional

ax.set(title='Contoh 3D Line Plot',
       xlabel='Sumbu X',
       ylabel='Sumbu Y',
       zlabel='Sumbu Z')

plt.show()
```

## 3D Scatter Plot


- Tiga variabel di bawah akan di plotting pada area 3D di atas dengan bentuk Scatter Plot

```python
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
```

```python
fig = plt.figure()             
ax = plt.axes(projection='3d')

ax.scatter(x, y, z, color='green')

ax.set(title='Contoh 3D Scatter Plot',
       xlabel='Sumbu X',
       ylabel='Sumbu Y',
       zlabel='Sumbu Z')

plt.show()
```

# #14: Contour Plot

- Membuat Contour Plot dengan Matplotlib

## Import modules

- `%matplotlib inline` untuk memanggil magic comment. Fungsinya untuk meredirect hasil output dari matplotlib ke tampilan jupyer   notebook di halaman notebook yang sama
- Kode di bawah ini untuk memberi tahu berapa versi matplotlib dan numpy kalian

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

## Sample dataset

- Fungsi `meshgrid()` untuk membentuk data area
- Ada tiga variabel, yaitu x, y, dan z 
- Tiga variabel tersebut akan mempresentasikan koordinat

```python
n = np.linspace(-3.0, 3.0, 100)

x, y = np.meshgrid(n, n) 
z = np.sqrt(x**2 + y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

## Contour Plot

- Contour plot merupakan visualisasi data yang bisa digunakan untuk memvisualisasikan data 3D ke dalam format 2D
- Perbedaan warna akan mempresentasikan perbedaan nilai

```python
plt.contour(x, y, z)
plt.colorbar()

plt.title('Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

## Filled Contour Plot

- Fungsi filled contour plot untuk memenuhi areanya dengan warna

```python
plt.contourf(x, y, z)
plt.colorbar()

plt.title('Filled Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

# #15: 3D Area Plotting

- Membuat 3D Area Plotting dengan Matplotlib

## Import Modules

- `%matplotlib inline` untuk memanggil magic comment. Fungsinya untuk meredirect hasil output dari matplotlib ke tampilan jupyer   notebook di halaman notebook yang sama
- Fungsi `%matplotlib notebook` sama seperti `%matplotlib inline`, tetapi sewaktu mengaktifkan `%matplotlib notebook` ini kita bisa berinteraksi dengan 3D plotting yang kita hasilkan
- Kode di bawah ini untuk memberi tahu berapa versi matplotlib dan numpy kalian

```python
# %matplotlib inline
%matplotlib notebook 
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

## Sample dataset

```python
n = np.linspace(-3.0, 3.0, 100)

x, y = np.meshgrid(n, n) 
z = np.sqrt(x**2 + y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

## 3D Wireframe

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_wireframe(x, y, z, color='orange')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Wireframe')

plt.show()
```

## 3D Surface Plot

- Fungsi `plot_surface` untuk membentuk 3D surface plot

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, z, color='skyblue')
# ax.plot_surface(x, y, z, cmap='viridis') # color map

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot')

plt.show()
```
