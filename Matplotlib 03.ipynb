## Simple Line Plot

- `'np.arange'` fungsinya adalah menyertakan tiga buah parameter
- Parameternya 0.0, 2.0, dan 0.01
- Parameter pertama mengindikasikan starting point atau nilai awalnya
- Parameter kedua mengintegrasikan nilai akhirnya
- Parameter ketiga menentukan interval nilainya
- Pada hasilnya bisa dilihat berupa deret bilangan yang ditampung dalam         struktur data array dan nilai awalnya adalah 0.0 lalu nilai akhirnya adalah   1.99
- Starting point untuk mengindikasikan nilai awal yang sifatnya disertakan     atau inclusive sedangkan nilai maksimumnya adalah 2.0 yang sifatnya           exclusive artinya hanya digunakan sebagi pembatas tetapi tidak disertakan     dalam deret bilangan


```python
x = np.arange(0.0, 2.0, 0.01)
x
```

- Membentuk dataset kedua yang akan ditampung dalam variabel 's'
- Pada dataset ini dasarnya untuk menampung nilai hasil operasi sin atau       hasil operasi sinus dari dataaet yang ditampung pada variabel x
- Fungsi sinus untuk sekumpulan data tertentu dan kumpulan datanya diambil     dari nilai x

```python
s = np.sin(2 * np.pi * x)
s
```

- `'plt.subplots()'` fungsinya untuk menghasilkan dua buah nilai yang perlu     ditampung dalam dua buah variabel
- Variabel pertama akan menampung figure dan variabel kedua akan menampung     aksesnya
- `'ax.plot'` fungsinya untuk melakukan plotting pada akses
- `'ax.grid()` fungsinya untuk menampilkan grid pada area plotting

```python
fig, ax = plt.subplots() # OO Style

ax.plot(x, s)
ax.set(xlabel='nilai x',
      ylabel='nilai sin(x)',
      title='Visualisasi nilai sin')
ax.grid()

plt.show()
```

- Plotting dengan cara pyplot style

```python
plt.plot(x,s)

plt.xlabel('nilai x')
plt.ylabel('nilai sin(x)')
plt.title('Visualisasi nilai sin')

plt.grid()
```

## Multiple Subplots

- Empat buah deret bilangan ditampung dalam empat variabel
- Nilai x1 dan x2 akan dihasilkan dengan pemanggilan fungsi `'np.linspace'`
- Parameter pertama digunakan untuk mengatur nilai terkecil atau nilai         minimum
- Parameter kedua digunakan untuk mengatur nilai terbesar atau nilai maksimum
- Parameter ketiga pada linspace mengindikasikan jumlah data point yang akan   digenerate
- Parameter minimum dan maksimum sifatnya inclusive berbeda dengan arange

```python
x1 = np.linspace(0.0, 5.0, 100)
x2 = np.linspace(0.0, 2.0, 100)

y1 = np.cos(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
x1
```

## Multiple Subplots dengan OO Style

- Dalam 1 figure terdapat 2 buah akses
- Saat memanggil `'plt.subplots'` akan membutuhkan dua buah variabel untuk menampung hasilnya
- Variabel pertama digunakan untuk menampung objek figure
- Variabel kedua didefinisikan tupple yang menampung dua buah akses
- Plotting pada akses pertama dan akses kedua

```python
fig, (ax1, ax2) = plt.subplots(2, 1) # membuat subplots
fig.suptitle('Multiple Subplots') # manampilkan judul 

# akses pertama
ax1.plot(x1, y1, 'ro-') # 'ro-' marker
ax1.set_ylabel('Nilai $cos(x_1)$')

# akses kedua
ax2.plot(x2, y2, 'g.-') # 'g.-' marker
ax2.set_ylabel('Nilai $cos(x_2)$')

ax2.set_xlabel('Nilai x')

plt.show()
```

## Multiple Subplots dengan pyplot Style

- Parameter pertama untuk mengatur layout

```python
# akses pertama
plt.subplot(2, 1, 1) # parameter pertama
plt.plot(x1, y1, 'ro-')
plt.title('Multiple Subplots')
plt.ylabel('Nilai $cos(x_1)$')

# akses kedua
plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'g.-')
plt.title('Nilai $cos(x_2)$')
plt.ylabel('Nilai $x$')

plt.show()
```

# Histogram

Cara membuat Histogram dengan matplotlib

## Import modules

- `'%matplotlib inline'`untuk memanggil magic comment. Fungsinya untuk         meredirect hasil output dari matplotlib ke tampilan jupyer notebook di       halaman notebook yang sama
- Kode di bawah ini untuk memberi tahu berapa versi matplotlib dan versi       numpy kalian

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

## Sample Data

- Variabel `'mu'` akan digunakan untuk merepresentasikan nilai mean
- Variabel `'sigma'` akan digunakan untuk merepresentasikan nilai standard     deviation 
- Nilai '10000' merepresentasikan jumlah data point yang akan dihasilkan oleh   fungsi `'np.random.randn'`
- `'np.random.randn'` fungsinya untuk menghasilkan sejumlah angka acak yang     akan membentuk distribution normal

```python
mu, sigma = 100, 15 

x = mu + sigma * np.random.randn(10000) # normal distribution
x 
```

```python
x.shape # melihat jumlah data yang ada di variabel x
```

## Histogram dengan pyplot Style

- `'plt.hist'` fungsinya untuk membangung histogram
- Pemanggilan `'plt.hist'` mangisyaratkan dua parameter yang harus disertakan
- `bins` digunakan untuk menentukan jumlah pengelompokan data

```python
plt.hist(x,  # sample data
        bins=50,  # parameter kedua
        facecolor='g', # warna green
        alpha=0.75) # mengatur transparansi

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Histogram')

plt.text(45, 500, r'$\mu=100,\ \sigma=15$')
plt.grid()

plt.show()
```

## Histogram dengan OO Style

- `'plt.hist'` fungsinya untuk membuat histogram

```python
fig, ax = plt.subplots()

ax.hist(x,  # sample data
        bins=50,  # parameter kedua
        facecolor='r', # warna red
        alpha=0.75) # mengatur transparansi

ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')
ax.set_title('Contoh Histogram')

ax.text(45, 500, r'$\mu=100,\ \sigma=15$') # data string
ax.grid()

plt.show()
```
