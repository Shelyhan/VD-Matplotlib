## Predefined Styles

- `plt.style.available` merupakan daftar dari Predefined Styles yang terdapat dalam matplotlib 

```python
plt.style.available
```

```python
plt.style.use('ggplot')
```

- Membentuk dataset untuk plotting
- `Variabel x` akan digunakan untuk menampung sekumpulan nilai
- `np.linspace` menghasilkan nilai

```python
x = np.linspace(1, 10, 15)

plt.plot(x, x, 'bs-', label='lin') # plotting pertama
plt.plot(x, np.log(x), 'ro-', label='log') # plotting kedua

plt.legend()
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Plot')

plt.show()
```

## Menerapkan Multiple Style

- Menerapkan lebih dari satu style sekaligus


```python
plt.style.use(['seaborn', 'dark_background'])
```

```python
x = np.linspace(1, 10, 15)

plt.plot(x, x, 'bs-', label='lin')
plt.plot(x, np.log(x), 'ro-', label='log') 

plt.legend()
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Plot')

plt.show()
```

## Menerapkan Temporary Styling

- Melakukan plotting sederhana
- Menggunakan style `ggplot` secara global

```python
plt.style.use('ggplot')

plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
plt.show()
```

- Menerapkan style `dark_background` yang sifatnya temporary

```python
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
    
plt.show()
```
