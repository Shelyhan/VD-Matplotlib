## Simple 2x2 layout dengan subplots

- Membuat plotting sederhana dengan layout 2x2 yang akan dibentuk dengan memanfaatkan subplots
- Menampung 4 buah object axes

```python
siswa = ['Farhan', 'Bagas', 'Putri', 'Clarisa']
uas   = [68, 83, 96, 72]

fig, axes = plt.subplots(ncols=2, nrows=2, constrained_layout=True) # 2x2 layout

axes[0][0].bar(siswa, uas)
axes[0][1].pie(uas, labels=siswa)
axes[1][0].pie(uas, labels=siswa, explode=[0, 0, 0, 0.2])
axes[1][1].barh(siswa, uas)

plt.show()
```

- Contoh tanpa plotting axes
- `constrained_layout=False` untuk merapatkan tampilan object dan akan saling tindih

## Simple 2x2 layout dengan GridSpec

- Membuat plotting sama seperti di atas tetapi menggunakan gridspec
- Gridspec mempunyai 2 cara
- Data di bawah menggunakan cara pertama gridspec 
- Figure di bawah dibentuk tidak menggunakan fungsi subplot, tetapi dengan memanggil `plt.figure`

```python
fig = plt.figure(constrained_layout=True)

gs  = gridspec.GridSpec(ncols=2, nrows=2, figure=fig)

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

ax1.bar(siswa, uas)
ax2.pie(uas, labels=siswa)
ax3.pie(uas, labels=siswa, explode=[0, 0, 0, 0.2])
ax4.barh(siswa, uas)

plt.show()
```

- Cara lain memanggil gridspec, yaitu dengan menggunakan fungsi `fig.add_gridspec`

```python
fig = plt.figure(constrained_layout=True)

gs  = fig.add_gridspec(ncols=2, nrows=2)

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

ax1.bar(siswa, uas)
ax2.pie(uas, labels=siswa)
ax3.pie(uas, labels=siswa, explode=[0, 0, 0, 0.2])
ax4.barh(siswa, uas)

plt.show()
```

## Rows Span dan Cols Span dengan GridSpec

- Memperluas kolom untuk baris pada index ke nol
- Nilai 0 mengindikasikan bahwa posisi barisnya berada pada index ke 0
- `:` artinya adalah menggunakan keseluruhan kolom yang tersedia

```python
fig = plt.figure(constrained_layout=True)

gs  = fig.add_gridspec(ncols=2, nrows=2)

ax1 = fig.add_subplot(gs[0, :])
# ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

ax1.bar(siswa, uas)
# ax2.pie(uas, labels=siswa)
ax3.pie(uas, labels=siswa, explode=[0, 0, 0, 0.2])
ax4.barh(siswa, uas)

plt.show()
```

- Layout dasar data di bawah adalah 3x3 sebelum dilakukan perluasan
- Axes pertama mengalami perluasan kolom karena menguasai semua kolom yang terdapat pada baris index ke `0`
- Axes kedua juga mengalami perluasan kolom dengan mengambil dua kolom
- Axes ketiga mengalami perluasan baris
- `-1` artinya menempati kolom terakhir

```python
fig = plt.figure(constrained_layout=True)
gs  = fig.add_gridspec(ncols=3, nrows=3)

ax1 = fig.add_subplot(gs[0, :])
ax1.set_title('gs[0, :]')

ax2 = fig.add_subplot(gs[1, :-1])
ax2.set_title('gs[1, :-1]')
ax2.barh(siswa, uas)

ax3 = fig.add_subplot(gs[1:, -1])
ax3.set_title('gs[1:, -1]')
ax3.bar(siswa, uas)

ax4 = fig.add_subplot(gs[-1, 0])
ax4.set_title('gs[-1, 0]')

ax5 = fig.add_subplot(gs[-1, -2])
ax5.set_title('gs[-1, -2]')

plt.show()
```
