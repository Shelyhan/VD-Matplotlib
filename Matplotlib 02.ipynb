## Import modules

```python
%matplotlib inline
# untuk memanggil magic comment
# fungsinya untuk meredirect hasil output dari matplotlib 
# ke tampilan jupyer notebook di halaman notebook yang sama
```

```python
import matplotlib # untuk mengimport module matplotlib
import matplotlib.pyplot as plt
import numpy as np # untuk mengimport module numpy

print(matplotlib.__version__)
print(np.__version__)
# code ini untuk memberi tahu berapa versi matplotlib dan versi numpy kalian
```

## Membuat plotting sederhana

```python
# Plotting sederhana
plt.plot([2, 5, 7, 11])
plt.ylabel('sumbu y')
plt.show() # fungsinya untuk membuat hasil tampilan menjadi lebih rapi tanpa
# munculnya tulisan-tulisan yang tidak terlalu dibutuhkan
```

 ● Pemanggilan fungsi plot( ) dapat dilakukan dengan menyertakan hanya sebuah      deret bilangan (list/array) ataupun dua buah deret bilangan (list/array).
 
 ● Pada kasus di atas termasuk deret bilangan list.
 
 ● Bila pemanggilan hanya menyertakan sebuah deret bilangan, maka nilai pada      deret bilangan tersebut akan dijadikan data point untuk nilai pada sumbu y;    sedangkan nilai pada sumbu x akan secara otomatis dibuatkan sebuah deret        bilangan terurut dengan nilai dimulai dari nol.

```python
plt.plot([2, 3, 4, 7], [5, 4, 2, 16]) 
# bilangan 2, 3, 4, 7 merepresentasikan deret yang akan berkorelasi dengan sumbu x
# bilangan 5, 4, 2, 16 merepresentasikan deret yang akan berkorelasi dengan sumbu y
plt.show()
```

## Pengaturan Format pada plot  

   ● Selain menyertakan dua buah deret bilangan sebagai parameter untuk sumbu        x dan y, terdapat parameter ketiga yang bisa kita sertakan untuk mengatur      warna dan jenis plotting.
   
   ● Pengaturan parameter ketiga ini sepenuhnya mengadopsi gaya formatting          pada MATLAB.
   
   ● Nilai default untuk parameter ketiga ini 'b-'. Arti dari 'b-' adalah b          berarti blue dan - (strip) berarti garis
   
   ● Pemanggilan fungsi axis() menyertakan sebuah list [xmin, xmax, ymin,ymax].

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') 
# 'ro' adalah parameter ketiga pengatur bentuk marker. 
# arti dari'ro' adalah r berarti red (merah) dan 'o' merepresentasikan markernya. 
# Jika kita ingin merubah warna titik atau markernya, misalnya warna blue dan 
# markernya bentuk bintang kita tinggal mengganti 'ro' menjadi 'b*'
plt.axis([0, 6, 0, 20]) 
# ini adalah list. Nilai keempat bilangan tersebut merepresentasikan 
# nilai xmin, xmax, ymin, dan ymax.

plt.show()
# di dalam tampilan markernya tidak lagi garis tetapi berupa . (titik)
```

### Multi plot dalam satu pemanggilan fungsi plot( )

```python
t = np.arange(0., 5., 0.2) 
# nilai 0 dan 5 merepresentasikan nilai min dan nilai max
# nilai 0.2 artinya di antara nilai 0 sampai nilai 5 ini terdapat
# sejumlah nilai dengan interval dan konsisten
t
```

```python
plt.plot(t, t, 'r--') 
# plot pertama menggunakan t dan t sebagai nilai x dan nilai y
# dan menggunakan 'r--' sebagai parameter ketiga
# arti dari '--' adalah garis putus putus dan 'r' artinya red
plt.plot(t, t**2, 'bs')
# plot kedua merepresentasikan data kuadrat dan 
# menggunakan t dan t sebagai nilai x dan nilai y
# 'bs' sebagai parameter ketiga. Arti dari 'bs' adalah 'b' merepresentasikan 
# warna biru lalu 's' merepresentasikan square (kotak)
plt.plot(t, t**3, 'g^')
# plot ketiga merupakan plotting cube dan data kuadrat
# menggunakan t dan t sebagai nilai x dan nilai y
# 'g^' sebagai parameter ketiga. Arti dari 'g^' adalah 'g' merepresentasikan
# green dan '^' merepresentasikan segitiga
plt.show()
```

Kita juga dapat melakukan multi plot dalam satu pemanggilan fungsi plot( )

```python
# melakukan multi plot
plt.plot(t, t, 'r--',
        t, t**2, 'bs',
        t, t**3, 'g^')
plt.show()
```

## Plotting dengan keywords

Matplotlib juga memungkinkan kita untuk melakukan plotting sekumpulan nilai yang tersimpan dalam struktur data yang disetai keywords argument (kwargs) seperti pada dictionary dan Pandas DataFrame.

```python
# melakukan multi plot
data = {'a' : np.arange(50),
        'c' : np.random.randint(0, 50, 50),
        'd' : np.random.randn(50)}
# dictionary yang cukup sederhana yang terdiri dari 3 buah key
data['b'] = data['a'] + 10 * np.random.randn(50) # key baru
data['d'] = np.abs(data['d']) * 100 # key baru

data
# pada hasil tampilan akan terlihat berbeda dengan milik kalian karna ini 
# melibatkan proses pengacakan
```

```python
# menggunakan tipe floating scatterplot
plt.scatter('a','b', data=data) 
# parameter terakhir adalah data yang kita ambil dari variabel data
plt.xlabel('entry a') # key 
plt.ylabel('entry b') # key
plt.show()
# scatterplot bisa menampilkan dimensi 3 dan dimensi 4 yang 
# nanti bisa di representasikan dengan warna dan ukuran
```

```python
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.xlabel('entry b')
plt.show()
# warna pada tampilan tergantung pada representasi dari 'c'
# ukuran pada tampilan tergantung pada representasi dari 'd'
```

## Plotting untuk tipe data categorial

Matplotlib juga mendukung plotting untuk data dengan tipe kategori atau biasa dikenal dengan istilah categorical data atau categorical variables.

```python
# dua data akan di plotting
names = ['group_a', 'group_b', 'group_c'] # variabel
values = [1, 10, 100] #variabel 
```

```python
plt.plot(names, values)
# names adalah sumbu x
# values adalah sumbu y
plt.show()
```

```python
plt.scatter(names, values)
plt.show()
```

```python
# tipe data bar plot
plt.bar(names, values)
plt.show()
```
