## Animasi pada Matplotlib

- `animate` fungsinya untuk menyiapkan nilai untuk x dan y yang akan dibentuk sekumpulan deret bilangan dengan nilai terkecil 0 dan nilai terbesar 4 sejumlah 1000 data
- `FuncAnimation` untuk menghasilkan object animasi
- `to_html5_video()` untuk objek animasi akan dikonversikan menjadi HTML5 video

```python
fig = plt.figure()
plt.xlim((0, 4))
plt.ylim((-2, 2))

line = plt.plot([])[0]

def animate(i):
    x = np.linspace(0, 4, 1000)
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x, y)
    return line,

anim = FuncAnimation(fig, animate, frames=200, interval=20)

video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)

plt.close()
```
